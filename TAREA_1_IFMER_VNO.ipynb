```

def seales(pixel_bytes):
  
  pixel_bytes.channels_no = pixel_bytes.NumberOfWaveformChannels
  pixel_bytes.channels_definitions=pixel_bytes.ChannelDefinitionSequence
  pixel_bytes.signals=pixel_bytes.values()
  pixel_bytes.samples= pixel_bytes.NumberOfWaveformSamples

  factor=np.zeros(pixel_bytes.channels_no)+1
  baseln=np.zeros(pixel_bytes.channels_no)
  units=[]
  for idx in range(pixel_bytes.channels_no):
    definition=pixel_bytes.channels_definitions[idx]

    assert(definition.WaveformBitsStored== 16)


    if definition.get('ChannelSensitivity'):
      factor[idx] = (
          float(definition.ChannelSensitivity) *
          float(definition.ChannelSensitivityCorrectionFactor)
      )


    if definition.get('ChannelBaseline'):
                baseln[idx] = float(definition.get('ChannelBaseline'))


    units.append(
                definition.ChannelSensitivityUnitsSequence[0].CodeValue
    )


  unpack_fmt= '<%dh' % (len(pixel_bytes.WaveformData) / 2)

  unpacked_waveform_data = struct.unpack(unpack_fmt, pixel_bytes.WaveformData)
  seales = np.asarray(
            unpacked_waveform_data,
            dtype=np.float64).reshape(
            pixel_bytes.samples,
            pixel_bytes.channels_no).transpose()
  for channel in range(pixel_bytes.channels_no):
    seales[channel] = (
        (seales[channel] + baseln[channel]) * factor[channel]
    )
  
  return seales

signalECG2= seales(pixel_bytes)
```

```
import sys
import argparse
from scp import *
from scpreader import FileReader, ScpReader
from scputil import ScpPrinter, lead_dic
from scpformat import format_section, format_samples_as_csv
```

```
def format_scp(f, csv_argument):
    fr = FileReader(f)
    scpReader = ScpReader(fr)
    #print(scpReader)
    scp = scpReader.read_scp()
    scpReader.close()
    printer = ScpPrinter()
    #print(printer)
    if csv_argument is not None:
        format_samples_as_csv(scp, csv_argument[0], printer)
        return

    printer.p('--ScpRec--', '----')
    printer.p('CRC', scp.crc)
    printer.p('RecLen', scp.len)

    for s in scp.sections:
        format_section(s, printer)


def argparser():
    parser = argparse.ArgumentParser()
    parser.add_argument('scpfile', help='input SCP file')
    parser.add_argument(
        '--csv', type=int, nargs=1, metavar='section_id', help='print leads in CSV format for section 5 or 6, specify here a section id')
    return parser
```
